```python
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

print(df.head())
print(df.shape)
print(df.dtypes)
```

```python
print("mean và Std theo sepal_length")
mean_value=df['sepal_length'].mean()
std_value=df['sepal_length'].std()
print("Mean:", mean_value)
print("Std:", std_value)

```

```python
normal = np.random.normal(loc=mean_value, scale=std_value, size=150)
print(normal)
print(normal.shape)
```

```python
sns.histplot(data=df, x='sepal_length', stat='density', color='skyblue', label='Thực tế')
x_axis = np.linspace(df['sepal_length'].min(), df['sepal_length'].max(), 100)
pdf_values = norm.pdf(x_axis, loc=mean_value, scale=std_value)
plt.plot(x_axis, pdf_values, color='red', linewidth=2, label='PDF Lý thuyết')
plt.legend()
plt.show()
```

```python
sns.histplot(x = normal, stat='density', color='skyblue', label='Dự đoán')
x_axis = np.linspace(df['sepal_length'].min(), df['sepal_length'].max(), 100)
pdf_values = norm.pdf(x_axis, loc=mean_value, scale=std_value)
plt.plot(x_axis, pdf_values, color='red', linewidth=2, label='PDF Lý thuyết')
plt.legend()
plt.show()
```

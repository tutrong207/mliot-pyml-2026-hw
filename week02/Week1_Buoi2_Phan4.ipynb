```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
P_B = 0.01
P_pos_givenB = 0.99
P_pos_givenNB = 0.05

P_pos = P_pos_givenB*P_B + P_pos_givenNB*(1 - P_B)
P_B_given_pos = P_pos_givenB * P_B / P_pos
print("P_B_given_pos =", round(P_B_given_pos,4))
```

```python
P_B_axis = np.linspace(0.001, 0.2, 100)
P_pos = P_pos_givenB*P_B_axis + P_pos_givenNB*(1 - P_B_axis)
P_B_given_pos = P_pos_givenB * P_B_axis / P_pos
plt.plot(P_B_axis, P_B_given_pos, color='red', linewidth=2, label='Đồ thị P(B|+) theo P(B)')
plt.legend()
plt.show()
```
